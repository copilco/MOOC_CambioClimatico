```python
#Cargamos las librerías
import numpy as np
import bokeh.plotting as bkp
import bokeh.models as bkm
import bokeh.io as bkio
from bokeh.colors import RGB
import pandas as pd
import datetime as dt

#Mostramos las gráficas dentro del notebook
bkp.output_notebook()
```

```python
#Cargamos los datos
afolu=pd.read_csv("FAOSTAT_data_5-7-2018.csv", sep=",", index_col="Year")
entfer=afolu[(afolu["Item"]=="Enteric Fermentation")]
manure=afolu[(afolu["Item"]=="Manure Management")]
rice=afolu[(afolu["Item"]=="Rice Cultivation")]
synthfer=afolu[(afolu["Item"]=="Synthetic Fertilizers")]
mansoil=afolu[(afolu["Item"]=="Manure applied to Soils")]
manpast=afolu[(afolu["Item"]=="Manure left on Pasture")]
cropres=afolu[(afolu["Item"]=="Crop Residues")]
culorg=afolu[(afolu["Item"]=="Cultivation of Organic Soils")]
burcrop=afolu[(afolu["Item"]=="Burning - Crop residues")]
bursav=afolu[(afolu["Item"]=="Burning - Savanna")]
```

```python
culorg.head(n=5)
```

```python
#Cargamos los colores elegidos como paleta para este MOOC
p_azulmarino = RGB(0, 28, 100)
p_azulclaro = RGB(0, 131, 255)
p_amarillo = RGB(244, 255, 84)
p_verde = RGB(172, 255, 165)
p_rojo = RGB(255, 67, 67)
```

```python
p = bkp.figure(tools=["pan","wheel_zoom","box_zoom","save,reset"], width=900, height=500,
               active_drag = "pan", active_scroll = "wheel_zoom")

año = entfer.index

banda_superior=entfer["Value"]
banda_inferior=np.zeros(len(año))
entfer_x = np.append(año, año[::-1])
entfer_y = np.append(banda_inferior, banda_superior[::-1])

p.patch(entfer_x, entfer_y, color=p_azulmarino, fill_alpha=0.6, line_alpha=0.3, legend="Fermentación Entérica")

banda_inferior = banda_superior
banda_superior = banda_superior + manure["Value"]
manure_x = np.append(año, año[::-1])
manure_y = np.append(banda_inferior, banda_superior[::-1])

p.patch(manure_x, manure_y, color=p_azulmarino, fill_alpha=1, line_alpha=0.3, legend="Gestión Estiércol")
```

```python
#Creamos una figura de Bokeh que va a contener nuestra gráfica. Indico el tamaño, que el eje X van a ser fechas
#y las herramientas activas (mover la gráfica y hacer zoom con el ratón)

p = bkp.figure(tools=["pan","wheel_zoom","box_zoom","save,reset"], width=900, height=500,
               active_drag = "pan", active_scroll = "wheel_zoom")

año = entfer.index

banda_superior=entfer["Value"]
banda_inferior=np.zeros(len(año))
entfer_x = np.append(año, año[::-1])
entfer_y = np.append(banda_inferior, banda_superior[::-1])

p.patch(entfer_x, entfer_y, color="darkgreen", fill_alpha=1, line_alpha=0.3, legend="Fermentación Entérica")

banda_inferior = banda_superior
banda_superior = banda_superior + manure["Value"]
manure_x = np.append(año, año[::-1])
manure_y = np.append(banda_inferior, banda_superior[::-1])

p.patch(manure_x, manure_y, color=p_azulmarino, fill_alpha=1, line_alpha=0.3, legend="Gestión Estiércol")

banda_inferior = banda_superior
banda_superior = banda_superior + manpast["Value"]
manpast_x = np.append(año, año[::-1])
manpast_y = np.append(banda_inferior, banda_superior[::-1])

p.patch(manpast_x, manpast_y, color=p_azulmarino, fill_alpha=0.6, line_alpha=0.3, legend="Estiércol en el Pasto")

banda_inferior = banda_superior
banda_superior = banda_superior + mansoil["Value"]
mansoil_x = np.append(año, año[::-1])
mansoil_y = np.append(banda_inferior, banda_superior[::-1])

p.patch(mansoil_x, mansoil_y, color=p_azulmarino, fill_alpha=0.3, line_alpha=0.3, legend="Estiércol para Fertilizar")

banda_inferior = banda_superior
banda_superior = banda_superior + synthfer["Value"]
synthfer_x = np.append(año, año[::-1])
synthfer_y = np.append(banda_inferior, banda_superior[::-1])

p.patch(synthfer_x, synthfer_y, color=p_rojo, fill_alpha=1, line_alpha=0.3, legend="Fertilizantes Sintéticos")


banda_inferior = banda_superior
banda_superior = banda_superior + rice["Value"]
rice_x = np.append(año, año[::-1])
rice_y = np.append(banda_inferior, banda_superior[::-1])

p.patch(rice_x, rice_y, color=p_verde, fill_alpha=1, line_alpha=0.3, legend="Cultivo de Arroz")

banda_inferior = banda_superior 
banda_superior = banda_superior + cropres["Value"]
cropres_x = np.append(año, año[::-1])
cropres_y = np.append(banda_inferior, banda_superior[::-1])

p.patch(cropres_x, cropres_y, color="yellow", fill_alpha=1, line_alpha=0.3, legend="Residuos de Cultivos")

banda_inferior = banda_superior
banda_superior = banda_superior + burcrop["Value"]
burcrop_x = np.append(año, año[::-1])
burcrop_y = np.append(banda_inferior, banda_superior[::-1])

p.patch(burcrop_x, burcrop_y, color="darkred", fill_alpha=0.5, line_alpha=0.3, legend="Quema de Residuos de Cultivos")

banda_inferior = banda_superior + culorg["Value"]
banda_inferior = banda_inferior - culorg["Value"]
banda_superior = banda_superior + culorg["Value"]
culorg_x = np.append(año, año[::-1])
culorg_y = np.append(banda_inferior, banda_superior[::-1])

p.patch(culorg_x, culorg_y, color=p_azulclaro, fill_alpha=1, line_alpha=0.3, legend="Cultivo de Suelos Orgánicos")


banda_inferior = banda_superior
banda_superior = banda_superior + bursav["Value"]
bursav_x = np.append(año, año[::-1])
bursav_y = np.append(banda_inferior, banda_superior[::-1])

p.patch(bursav_x, bursav_y, color="darkred", fill_alpha=1, line_alpha=0.3, legend="Quema de la Savana")
```

```python
#Indico las propiedades del gráfico
p.title.text = "Emisiones sector AFOLU"
p.title.text_color = "black"
p.title.text_font = "ADAM CG PRO"
p.title.text_font_style = "normal"
p.title.text_font_size = "16pt"
p.title.align = "center"

p.background_fill_color = "white"
p.background_fill_alpha = 1.

p.xaxis.axis_label = 'Año'
p.xaxis.axis_label_text_font = "ADAM CG PRO"
p.xaxis.axis_label_text_font_style  = "normal"
p.xaxis.axis_label_text_font_size = "12pt"
p.yaxis.axis_label = 'Miles de Toneladas eq. CO2'
p.yaxis.axis_label_text_font = "ADAM CG PRO"
p.yaxis.axis_label_text_font_style  = "normal"
p.yaxis.axis_label_text_font_size = "12pt"
p.yaxis[0].formatter = bkm.NumeralTickFormatter(format="0,0")

p.y_range = bkm.Range1d(start=0, end=10000000)

p.ygrid.grid_line_color=p_azulclaro
p.ygrid.grid_line_alpha=0.25
p.xgrid.grid_line_color=p_azulclaro
p.xgrid.grid_line_alpha=0.0

p.xaxis.major_tick_line_color = p_rojo
p.xaxis.major_tick_line_width = 3
p.xaxis.minor_tick_line_color = p_rojo
p.xaxis.minor_tick_line_width = 1
p.yaxis.major_tick_line_color = p_rojo
p.yaxis.major_tick_line_width = 3
p.yaxis.minor_tick_line_color = p_rojo
p.yaxis.minor_tick_line_width = 1

p.outline_line_width = 7
p.outline_line_alpha = 0.5
p.outline_line_color = p_azulclaro

p.legend.location = "top_left"
p.legend.click_policy="hide"
p.legend.label_text_font = "ADAM CG PRO"
```
